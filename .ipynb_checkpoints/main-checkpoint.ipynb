```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mpl_toolkits
%matplotlib inline
```

```python
data = pd.read_csv("kc_house_data.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
sns.set(font_scale=1.4)
data['bedrooms'].value_counts().plot(kind='bar',figsize=(7, 6), rot=0)
plt.title('number of Bedroom')
plt.xlabel('Bedrooms')
plt.ylabel('Count')
sns.despine
```

```python
plt.figure(figsize=(10,10))
sns.jointplot(x=data.lat.values, y=data.long.values, height=10)
plt.ylabel('Longitude', fontsize=12)
plt.xlabel('Latitude', fontsize=12)
plt.show()
sns.despine
```

```python
plt.scatter(data.price,data.sqft_living)
plt.title("Price vs Square Feet")
plt.ylabel('Square Feet', fontsize=15)
plt.xlabel('Price', fontsize=15)
plt.show()
```

```python
plt.scatter(data.price,data.long)
plt.title("Price vs Location of the area")
plt.ylabel('Location of the area', fontsize=15)
plt.xlabel('Price', fontsize=15)
plt.show()
```

```python
plt.scatter(data.price,data.lat)
plt.xlabel("Price")
plt.ylabel('Latitude')
plt.title("Latitude vs Price")
```

```python
plt.scatter(data.bedrooms,data.price)
plt.title("Bedroom and Price ")
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.show()
sns.despine
```

```python
plt.scatter((data['sqft_living']+data['sqft_basement']),data['price'])
plt.xlabel("Living and basement square feet area")
plt.ylabel("Price")
plt.show()
sns.despine
```

```python
plt.scatter(data.waterfront,data.price)
plt.title("Waterfront vs Price ( 0= no waterfront)")
```
